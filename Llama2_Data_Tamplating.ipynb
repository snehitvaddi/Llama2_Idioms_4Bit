### Idioms Customs Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import random
from tqdm import tqdm

# Load the dataset from an Excel file
df = pd.read_excel('Idioms_Context.xlsx', engine='openpyxl')

# Define a function to create various response formats for the idiom
def create_idiom_response(row):
    idiom = row['Idiom']
    context = row['Example']
    response_templates = [
        f"The idiom '{idiom}' is used.",
        f"In this sentence, the idiom '{idiom}' is illustrated.",
        f"Here, '{idiom}' is the expression in play.",
        f"This example showcases the idiom '{idiom}'.",
        f"Idiom present: '{idiom}'.",
    ]
    # Randomly select a template for the response
    response = random.choice(response_templates)
    response = response.replace('"', '')
    context = context.replace('"', '')
    return f"\"<s>[INST] {context} [/INST] {response}</s>\""

# Register pandas progress_apply with tqdm
tqdm.pandas()

# Apply the function with a progress bar
df['transformed'] = df.progress_apply(create_idiom_response, axis=1)

# Split the data into train and test sets
train_df, test_df = train_test_split(df['transformed'], test_size=0.2, random_state=42)

# Write the train and test sets to CSV files
train_df.to_csv('train.csv', index=False, header="Prompt")
test_df.to_csv('test.csv', index=False, header="Prompt")
```
